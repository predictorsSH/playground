```python
my_dict = {
    'a': 1,
    'b': 2,
    'c': 3
}


# code smell
for key in my_dict.keys():
    print(key, my_dict[key])

# better
for key in my_dict:
    print(key, my_dict[key])

# better
for key, value in my_dict.items():
    print(key, value)
```
